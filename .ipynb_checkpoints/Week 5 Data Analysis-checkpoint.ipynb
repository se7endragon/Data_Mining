```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV
from category_encoders import *
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from scipy import stats
from sklearn import svm
df = pd.read_excel('Week 3 Data/CallsData.xls')
df2 = pd.read_csv('Week 3 Data/ContractData.csv')
#inner join (join on index should be used by pd.join(how='inner'))
df = pd.merge(df,df2,on=['Area Code','Phone'])
#column filter
df.drop(['Phone','Account Length'],axis=1,inplace=True)
df_logistic = df.drop(['State'],axis=1)
#number to string(categorical)
print(df['Churn'].unique()) #its binary, so we dont have to encode this one.
print(df['Area Code'].unique())
print(df['State'].unique())
enc = BinaryEncoder(cols=['Area Code','State']).fit(df)
enc_df = enc.transform(df)
enc_logistic = BinaryEncoder(cols=['Area Code']).fit(df_logistic)
enc_df_logistic = enc_logistic.transform(df_logistic)
#normalize using z-score
enc_df = enc_df.apply(lambda x: x if np.std(x) == 0 or x.name == 'Churn' or 'Area Code' in x.name or 'State' in x.name else stats.zscore(x))
enc_df_logistic = enc_df_logistic.apply(lambda x: x if np.std(x) == 0 or x.name == 'Churn' or 'Area Code' in x.name else stats.zscore(x))
```

```python
#Partition
X = enc_df.loc[:,enc_df.columns != 'Churn'].values
y = enc_df.loc[:,enc_df.columns == 'Churn'].values.ravel()
```

```python
X[0]
```

```python
skf = StratifiedKFold(n_splits=5)
sm = SMOTE()
classifier = DecisionTreeClassifier(criterion='gini',random_state=0,min_samples_split=6)
```

```python
#cross-validation: Gini-index
gini_classifier = DecisionTreeClassifier(criterion='gini',random_state=0,min_samples_split=6)
for train_index, test_index in skf.split(X, y):
    X_train,X_test = X[train_index], X[test_index]
    y_train,y_test = y[train_index], y[test_index]
    X_res, y_res = sm.fit_resample(X_train,y_train.ravel())    
    gini_classifier.fit(X_res,y_res)
    y_pred = gini_classifier.predict(X_test)
    cm = confusion_matrix(y_test,y_pred)
    print("Feature_importances:", gini_classifier.feature_importances_,"len_TRAIN:", len(train_index), "len_TEST:", len(test_index), "Confusion Matrix:", cm, "Accuracy:", gini_classifier.score(X_test,y_test))
```

```python
#cross-validation: Gain-Ratio
gain_classifier = DecisionTreeClassifier(criterion='entropy',random_state=0,min_samples_split=6)
for train_index, test_index in skf.split(X, y):
    X_train,X_test = X[train_index], X[test_index]
    y_train,y_test = y[train_index], y[test_index]
    X_res, y_res = sm.fit_resample(X_train,y_train.ravel())    
    gain_classifier.fit(X_res,y_res)
    y_pred = gain_classifier.predict(X_test)
    cm = confusion_matrix(y_test,y_pred)
    print("Feature_importances:", gain_classifier.feature_importances_,"len_TRAIN:", len(train_index), "len_TEST:", len(test_index), "Confusion Matrix:", cm, "Accuracy:", gain_classifier.score(X_test,y_test))
```

```python
#cross-validation: Logistic-Regression
logistic_classifier = LogisticRegression(solver='sag',max_iter=4000)
#Partition
X_logit = enc_df_logistic.loc[:,enc_df_logistic.columns != 'Churn'].values
y_logit = enc_df_logistic.loc[:,enc_df_logistic.columns == 'Churn'].values.ravel()
for train_index, test_index in skf.split(X_logit, y_logit):
    X_train,X_test = X_logit[train_index], X_logit[test_index]
    y_train,y_test = y_logit[train_index], y_logit[test_index]  
    logistic_classifier.fit(X_train,y_train)
    y_pred = logistic_classifier.predict(X_test)
    cm = confusion_matrix(y_test,y_pred)
    print("len_TRAIN:", len(train_index), "len_TEST:", len(test_index), "Confusion Matrix:", cm, "Accuracy:", logistic_classifier.score(X_test,y_test))
```

```python
#cross-validation: SVM
SVM_classifier = svm.SVC(C=1,gamma=0.1)
for train_index, test_index in skf.split(X, y):
    X_train,X_test = X[train_index], X[test_index]
    y_train,y_test = y[train_index], y[test_index]  
    SVM_classifier.fit(X_train,y_train)
    y_pred = SVM_classifier.predict(X_test)
    cm = confusion_matrix(y_test,y_pred)
    print("len_TRAIN:", len(train_index), "len_TEST:", len(test_index), "Confusion Matrix:", cm, "Accuracy:", SVM_classifier.score(X_test,y_test))
```

# 5-2

```python
df = pd.read_csv("Week 3 Data/BUYTEST.csv")
#Converting strange null to np null
compare_str = df['AGE'][9]
for col in df.columns:
    df.loc[df[col] == compare_str, col] = np.nan
#Type conversion of columns
df[['AGE','INCOME','MARRIED','FICO','OWNHOME']] = df[['AGE','INCOME','MARRIED','FICO','OWNHOME']].apply(pd.to_numeric)
#Missing Value treatment
df['AGE'].fillna(df['AGE'].mean(),inplace=True)
df['FICO'].fillna(df['FICO'].mean(),inplace=True)
df['INCOME'].fillna(df['INCOME'].mean(),inplace=True)
df['SEX'].fillna(df['SEX'].value_counts().idxmax(),inplace=True)
df['MARRIED'].fillna(df['MARRIED'].value_counts().idxmax(),inplace=True)
df['OWNHOME'].fillna(df['OWNHOME'].value_counts().idxmax(),inplace=True)
df['ORGSRC'].fillna(df['ORGSRC'].value_counts().idxmax(),inplace=True)
#Categorical Encoder
enc = BinaryEncoder(cols=['SEX', 'LOC','ORGSRC']).fit(df)
enc_df = enc.transform(df)
enc_df.drop(['ID','C1','C2','C3','C4','C5','C6','C7','PURCHTOT'],axis=1,inplace=True)
```

```python
X = enc_df.loc[:,enc_df.columns != 'RESPOND'].values
y = enc_df.loc[:,enc_df.columns == 'RESPOND'].values
```

```python
sm = SMOTE()
kf = StratifiedKFold(n_splits = 10)
dt = DecisionTreeClassifier()
pipeline = Pipeline([('sm', sm), ('dt', dt)])
params = {'dt__min_samples_split': [2,6,13,65,129,259,388,646,1293],
          'dt__criterion': ['gini','entropy']}
grid = GridSearchCV(pipeline, params, cv=kf)
grid.fit(X, y)
```

```python
#sorted(pipeline.get_params().keys()) <- 이걸로 estimator param 체크
grid.best_params_
```

```python
grid.best_score_
```

```python
df = pd.read_excel('Week 3 Data/CallsData.xls')
df2 = pd.read_csv('Week 3 Data/ContractData.csv')
#inner join (join on index should be used by pd.join(how='inner'))
df = pd.merge(df,df2,on=['Area Code','Phone'])
#column filter
df.drop(['Phone','Account Length'],axis=1,inplace=True)
enc = BinaryEncoder(cols=['Area Code','State']).fit(df)
enc_df = enc.transform(df)
```

```python
from sklearn.feature_selection import RFECV
from sklearn.svm import SVR
```

```python
enc_df.columns
```

```python
#normalize using z-score
enc_df = enc_df.apply(lambda x: x if np.std(x) == 0 or x.name == 'Churn' or 'Area Code' in x.name or 'State' in x.name else stats.zscore(x))
```

```python
enc_df.describe()
```

```python
#Partition
X = enc_df.loc[:,enc_df.columns != 'Churn'].values
y = enc_df.loc[:,enc_df.columns == 'Churn'].values.ravel()
```

```python
#RFE는 SVR의 linear 커널만 사용가능. 메뉴얼한 버젼은 MLSeminar 리포에 있긴 하기때문에 여기선 sklearn에서 제공하는 RFE를 사용하는 목적이 크므로.
estimator = LogisticRegression()
selector = RFECV(estimator, step=1,cv=5)
selector = selector.fit(X, y)
```

```python
X_masked = enc_df.loc[:,selector.support_].values
```
