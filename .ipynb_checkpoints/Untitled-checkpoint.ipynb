```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
d = {'Fold':[1,2,3,4,5],'Fold Accuracy': [100-14.543,100-14.543,100-14.543,100-14.414,100-14.414]}
data = pd.DataFrame(d)
```

```python
data['Fold Accuracy'].describe()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(15,10))
ax = sns.barplot(x='Fold', y='Fold Accuracy', color="salmon",data=data, saturation=.5)
ax.set_ylim([80,95])
ax.axhline(y=data['Fold Accuracy'].mean())
ax.set_label('hi')
```
